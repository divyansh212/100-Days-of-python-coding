```python
Menu = {
    "espresso": {
        "Ingredients": {
            "Water": 50,
            "Coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "Ingredients": {
            "Water": 200,
            "Milk": 150,
            "Coffee": 24,
        },
        "cost": 2.50,
    },
    "cappuccino": {
        "Ingredients": {
            "Water": 250,
            "Milk": 100,
            "Coffee": 24,
        },
        "cost": 3.00,
    }
}

RESOURCES = {
    "Water": 350,
    "Milk": 200,
    "Coffee": 100
}

def payment_system(type_of_coffee):
    coffee_price = Menu[type_of_coffee]["cost"]

    PP = int(input("Enter the amount of Penny: "))
    PN = int(input("Enter the amount of Nickel: "))
    PD = int(input("Enter the amount of Dime: "))
    PQ = int(input("Enter the amount of Quarter: "))

    user_money = PP * 0.01 + PN * 0.05 + PD * 0.1 + PQ * 0.25
    change = user_money - coffee_price

    if user_money >= coffee_price:
        print(f"Payment successful! Here is your change: ${change:.2f}")
        print("Your coffee will be ready in a moment ☕")
        return True  # Payment successful
    else:
        print(f"PAYMENT UNSUCCESSFUL. Please add ${-change:.2f} more to complete the order.")
        return False  # Payment failed

def resource_checking(type_of_coffee):
    global RESOURCES

    ingredients = Menu[type_of_coffee]["Ingredients"]

    # Check if resources are sufficient
    for item in ingredients:
        if RESOURCES[item] < ingredients[item]:
            print(f"Not enough {item}. Please restock!")
            return False  # Insufficient resources

    # Deduct resources
    for item in ingredients:
        RESOURCES[item] -= ingredients[item]

    return True  # Resources are sufficient

def order_taking(coffee):
    if coffee == "report":
        print(RESOURCES)
        return

    if coffee not in Menu:
        print("Invalid choice. Please choose Espresso, Latte, or Cappuccino.")
        return

    if resource_checking(coffee):
        if payment_system(coffee):
            print(f"Here is your {coffee.capitalize()} ☕. Thank you for ordering!")
        else:
            print("Order canceled due to insufficient payment.")
    else:
        print("Order canceled due to insufficient resources.")
is_on=True
while is_on:
  coffee = input("What kind of coffee do you want (Espresso/Latte/Cappuccino)?\n").lower()
  order_taking(coffee)
  if coffee=="off":
    is_on=False
```

```python
import turtle
from PIL import Image
import os

# Create a turtle screen
screen = turtle.Screen()
screen.bgcolor("white")

# Create a turtle object
tom = turtle.Turtle()
tom.shape("turtle")
tom.color("blue")
tom.pensize(3)
tom.speed(3)

# Draw a square
for _ in range(4):
    tom.forward(100)
    tom.right(90)

# Save the drawing as an EPS file
eps_file = "turtle_drawing.eps"
png_file = "turtle_drawing.png"
ts = tom.getscreen()
ts.getcanvas().postscript(file=eps_file)

# Convert EPS to PNG using Ghostscript
os.system(f"gs -o {png_file} -sDEVICE=pngalpha {eps_file}")

# Display the image in Colab
from IPython.display import display
img = Image.open(png_file)
display(img)

# Close the turtle window
turtle.bye()
```
